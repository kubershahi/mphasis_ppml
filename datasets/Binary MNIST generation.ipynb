```python
import pandas as pd

df = pd.read_csv (r'mnist/mnist_test.csv')
print (df)
```

```python
labels = df.iloc[: , -1]
labels.nunique()
```

```python
for i in range(0,len(labels)):
    if labels[i] != 0:
        labels[i] = 1
```

```python
labels.nunique()
```

```python
df.iloc[: , -1] = labels
```

```python
df
```

```python
df.to_csv('binary_mnist/mnist_test.csv', index=False) 
```
